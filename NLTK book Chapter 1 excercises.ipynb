27.Define a function called vocab_size(text) that has a single parameter for the text, and which returns the vocabulary size of the text.

```python
import string

def vocab_size(text):
    if isinstance(text, str) == True: #if its a string (and not a list)
        text = text.translate(str.maketrans('', '', string.punctuation)) ##remove any punctuation from string
        text = text.split()        #split string into list
    return len(set(word.lower() for word in text if not word.isdigit()))
```

```python
sent5 = ["I", "cat", "car"]

vocab_size(sent5)
```

28.Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.

```python
def percent(word, text):
    word = word.lower()
    if isinstance(text, str) == True: #if its a string (and not a list)
        text = text.translate(str.maketrans('', '', string.punctuation)) ##remove any punctuation from string
        text = text.split()        #split string into list
    for i in range(len(text)):
        text[i] = text[i].lower()
    return(text.count(word)/len(list(w for w in text if w.isalpha())))*100
    
```
